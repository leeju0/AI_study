```python
import cv2 #Computer Vision의 약자 버전 2
import numpy as np
import matplotlib.pyplot as plt

cat1 = cv2.imread('/content/drive/MyDrive/리트리버.jpg')
plt.imshow(cat1)
plt.show()

cat1 = np.array(cat1)
print(cat1.shape)
```

```python
#R channel만 슬라이싱
red_cat = cat1[:,:,0]

#G 채널만 슬라이싱
green_cat = cat1[:,:,1]

#B 채널만 슬라이싱
blue_cat = cat1[:,:,2]


plt.imshow(red_cat,cmap='Reds'),
plt.title('R')
plt.show()

plt.imshow(green_cat,cmap='Greens'),
plt.title('G')
plt.show()

plt.imshow(blue_cat,cmap='Blues'),
plt.title('B')
plt.show()
```
